```python
#Scraping a specific user's tweets and placing them into a json file
!snscrape --jsonl --max-results 10 twitter-user ftab0007 > text_files_test_10.json

import json
import pandas as pd

table = []
with open('text_files_test_10.json', 'r') as f:
    for line in f:
        table.append(json.loads(line))

    pd.DataFrame.from_records(table).to_csv('text_files_test_10.csv')
    
data = pd.DataFrame.from_records(table)

data
```
